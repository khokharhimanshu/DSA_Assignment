```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def convertToBST(root):
    if not root:
        return None

    values = []
    inorderTraversal(root, values)

    new_root = constructBSTFromSortedArray(values)

    return new_root

def inorderTraversal(node, values):
    if node:
        inorderTraversal(node.left, values)
        values.append(node.val)
        inorderTraversal(node.right, values)

def constructBSTFromSortedArray(values):
    if not values:
        return None

    mid = len(values) // 2

    root = TreeNode(values[mid])
    root.left = constructBSTFromSortedArray(values[:mid])
    root.right = constructBSTFromSortedArray(values[mid+1:])

    return root

def inorderTraversalPrint(node):
    if node:
        inorderTraversalPrint(node.left)
        print(node.val, end=" ")
        inorderTraversalPrint(node.right)

root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

print("Binary Tree (before conversion):")
inorderTraversalPrint(root)
print()

new_root = convertToBST(root)

print("Binary Search Tree (after conversion):")
inorderTraversalPrint(new_root)
print()
```

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def findDistanceBST(root, node1, node2):
    path1 = findPath(root, node1)
    path2 = findPath(root, node2)

    lca_val = findLCAValue(path1, path2)
    lca = findNode(root, lca_val)

    distance1 = findDistanceFromNode(lca, node1)
    distance2 = findDistanceFromNode(lca, node2)

    return distance1 + distance2

def findPath(node, target):
    if node is None:
        return []

    if node.val == target:
        return [node.val]

    left_path = findPath(node.left, target)
    if left_path:
        return [node.val] + left_path

    right_path = findPath(node.right, target)
    if right_path:
        return [node.val] + right_path

    return []

def findLCAValue(path1, path2):
    lca_val = None

    for val1, val2 in zip(path1, path2):
        if val1 == val2:
            lca_val = val1
        else:
            break

    return lca_val

def findNode(node, target):
    if node is None or node.val == target:
        return node

    if target < node.val:
        return findNode(node.left, target)
    else:
        return findNode(node.right, target)

def findDistanceFromNode(node, target):
    if node is None:
        return 0

    if node.val == target:
        return 0

    if target < node.val:
        return 1 + findDistanceFromNode(node.left, target)
    else:
        return 1 + findDistanceFromNode(node.right, target)

# Example usage:
root = TreeNode(8)
root.left = TreeNode(3)
root.left.left = TreeNode(1)
root.left.right = TreeNode(6)
root.left.right.left = TreeNode(4)
root.left.right.right = TreeNode(7)
root.right = TreeNode(10)
root.right.right = TreeNode(14)
root.right.right.left = TreeNode(13)

node1 = 6
node2 = 14
distance1 = findDistanceBST(root, node1, node2)
print(f"The distance between node {node1} and node {node2} = {distance1}")

node3 = 3
node4 = 4
distance2 = findDistanceBST(root, node3, node4)
print(f"The distance between node {node3} and node {node4} = {distance2}")
```

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class DoublyLinkedListNode:
    def __init__(self, val):
        self.val = val
        self.prev = None
        self.next = None

def binaryTreeToDoublyLinkedList(root):
    if not root:
        return None

    head, tail = convert(root)

    return head

def convert(node):
    if not node:
        return None, None

    left_head, left_tail = convert(node.left)
    right_head, right_tail = convert(node.right)

    current = DoublyLinkedListNode(node.val)

    if left_tail:
        left_tail.next = current
        current.prev = left_tail
    else:
        left_head = current

    if right_head:
        right_head.prev = current
        current.next = right_head
    else:
        right_tail = current

    return left_head, right_tail

root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(20)
root.right.left = TreeNode(30)
root.right.right = TreeNode(35)

doubly_linked_list_head = binaryTreeToDoublyLinkedList(root)

current = doubly_linked_list_head
while current:
    print(current.val, end=" ")
    current = current.next
```
