```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverse_linked_list(head):
    prev = None
    current = head

    while current is not None:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node

    return prev

def is_palindrome(head):
    if head is None or head.next is None:
        return True 

    slow_ptr = head
    fast_ptr = head
    prev_ptr = None

    while fast_ptr is not None and fast_ptr.next is not None:
        fast_ptr = fast_ptr.next.next
        prev_ptr = slow_ptr
        slow_ptr = slow_ptr.next

    if fast_ptr is not None:
 
        slow_ptr = slow_ptr.next

    reversed_head = reverse_linked_list(slow_ptr)

    first_half = head
    second_half = reversed_head

    while second_half is not None:
        if first_half.data != second_half.data:
            return False  
        first_half = first_half.next
        second_half = second_half.next

    return True  

head1 = Node('R')
second1 = Node('A')
third1 = Node('D')
fourth1 = Node('A')
fifth1 = Node('R')
head1.next = second1
second1.next = third1
third1.next = fourth1
fourth1.next = fifth1

is_palindrome1 = is_palindrome(head1)
print(is_palindrome1)

head2 = Node('C')
second2 = Node('O')
third2 = Node('D')
fourth2 = Node('E')
head2.next = second2
second2.next = third2
third2.next = fourth2

is_palindrome2 = is_palindrome(head2)
print(is_palindrome2)
```
