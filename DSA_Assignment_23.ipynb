```python
def find_depth(preorder):
    def helper(preorder, index):
        if index >= len(preorder) or preorder[index] == 'l':
            return 0, index

        left_depth, index = helper(preorder, index + 1)
        right_depth, index = helper(preorder, index + 1)

        return 1 + max(left_depth, right_depth), index

    depth, _ = helper(preorder, 0)
    return depth

preorder1 = "nlnll"
print(find_depth(preorder1))  

preorder2 = "nlnnlll"
print(find_depth(preorder2))  
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def print_left_view(root):
    if root is None:
        return

    queue = []
    queue.append(root)
    left_view = []

    while queue:
        size = len(queue)

        for i in range(size):
            node = queue.pop(0)

            if i == 0:
                left_view.append(node.data)

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

    for node in left_view:
        print(node, end=" ")


root1 = Node(4)
root1.left = Node(5)
root1.right = Node(2)
root1.right.left = Node(3)
root1.right.right = Node(1)
root1.right.left.left = Node(6)
root1.right.left.right = Node(7)

print("Left View 1:")
print_left_view(root1)
print()

root2 = Node(1)
root2.left = Node(2)
root2.right = Node(3)
root2.left.right = Node(4)
root2.left.right.right = Node(5)
root2.left.right.right.right = Node(6)

print("Left View 2:")
print_left_view(root2)
print()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def print_right_view(root):
    if root is None:
        return

    queue = []
    queue.append(root)

    while queue:
        size = len(queue)

        for i in range(size):
            node = queue.pop(0)

            if i == size - 1:
                print(node.data, end=" ")

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)


root1 = Node(1)
root1.left = Node(2)
root1.right = Node(3)
root1.left.left = Node(4)
root1.left.right = Node(5)
root1.right.left = Node(6)
root1.right.right = Node(7)
root1.right.right.right = Node(8)

print("Right View 1:")
print_right_view(root1)
print()

root2 = Node(1)
root2.left = Node(8)
root2.left.left = Node(7)

print("Right View 2:")
print_right_view(root2)
print()
```

```python
from collections import deque, defaultdict

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        self.hd = 0


def print_bottom_view(root):
    if root is None:
        return

    bottom_view_dict = defaultdict(Node)

    queue = deque()
    queue.append(root)

    while queue:
        node = queue.popleft()

        hd = node.hd
        bottom_view_dict[hd] = node

        if node.left:
            node.left.hd = hd - 1
            queue.append(node.left)
        if node.right:
            node.right.hd = hd + 1
            queue.append(node.right)

    for key in sorted(bottom_view_dict):
        print(bottom_view_dict[key].data, end=" ")


root1 = Node(20)
root1.left = Node(8)
root1.right = Node(22)
root1.left.left = Node(5)
root1.left.right = Node(3)
root1.right.right = Node(25)
root1.left.right.left = Node(10)
root1.left.right.right = Node(14)

print("Bottom View 1:")
print_bottom_view(root1)
print()

root2 = Node(20)
root2.left = Node(8)
root2.right = Node(22)
root2.left.left = Node(5)
root2.left.right = Node(3)
root2.right.left = Node(4)
root2.right.right = Node(25)
root2.left.right.left = Node(10)
root2.left.right.right = Node(14)

print("Bottom View 2:")
print_bottom_view(root2)
print()
```
