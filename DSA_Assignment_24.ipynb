```python
def roman_to_int(s):
    roman_values = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
    total = 0
    prev_value = 0

    for i in range(len(s)):
        current_value = roman_values[s[i]]
        if current_value > prev_value:
            total += current_value - 2 * prev_value
        else:
            total += current_value
        prev_value = current_value

    return total


s1 = "III"
print("Roman numeral:", s1)
print("Integer value:", roman_to_int(s1))
print()

s2 = "LVIII"
print("Roman numeral:", s2)
print("Integer value:", roman_to_int(s2))
print()
```

```python
def length_of_longest_substring(s):
    unique_chars = set()

    start = 0
    end = 0

    max_length = 0

    while end < len(s):
        if s[end] not in unique_chars:
            unique_chars.add(s[end])
            max_length = max(max_length, end - start + 1)
            end += 1
        else:
            unique_chars.remove(s[start])
            start += 1

    return max_length


s1 = "abcabcbb"
print("Input:", s1)
print("Length of longest substring without repeating characters:", length_of_longest_substring(s1))
print()

s2 = "bbbbb"
print("Input:", s2)
print("Length of longest substring without repeating characters:", length_of_longest_substring(s2))
print()

s3 = "pwwkew"
print("Input:", s3)
print("Length of longest substring without repeating characters:", length_of_longest_substring(s3))
print()
```

```python
def majority_element(nums):
    count = 0
    candidate = None

    for num in nums:
        if count == 0:
            candidate = num
        count += 1 if num == candidate else -1

    count = 0
    for num in nums:
        if num == candidate:
            count += 1

    if count > len(nums) // 2:
        return candidate
    else:
        return None


nums1 = [3, 2, 3]
print("Input:", nums1)
print("Majority Element:", majority_element(nums1))
print()

nums2 = [2, 2, 1, 1, 1, 2, 2]
print("Input:", nums2)
print("Majority Element:", majority_element(nums2))
print()
```

```python
def group_anagrams(strs):
    anagrams = {}
    
    for s in strs:
        sorted_str = ''.join(sorted(s))
        if sorted_str in anagrams:
            anagrams[sorted_str].append(s)
        else:
            anagrams[sorted_str] = [s]
    
    result = list(anagrams.values())
    return result


strs1 = ["eat", "tea", "tan", "ate", "nat", "bat"]
print("Input:", strs1)
print("Grouped Anagrams:", group_anagrams(strs1))
print()

strs2 = [""]
print("Input:", strs2)
print("Grouped Anagrams:", group_anagrams(strs2))
print()

strs3 = ["a"]
print("Input:", strs3)
print("Grouped Anagrams:", group_anagrams(strs3))
print()
```

```python
def nth_ugly_number(n):
    ugly = [0] * n
    ugly[0] = 1
    p2 = p3 = p5 = 0

    for i in range(1, n):
        next_ugly = min(ugly[p2] * 2, ugly[p3] * 3, ugly[p5] * 5)
        ugly[i] = next_ugly

        if next_ugly == ugly[p2] * 2:
            p2 += 1
        if next_ugly == ugly[p3] * 3:
            p3 += 1
        if next_ugly == ugly[p5] * 5:
            p5 += 1

    return ugly[n - 1]


n1 = 10
print("Input:", n1)
print("Nth Ugly Number:", nth_ugly_number(n1))
print()

n2 = 1
print("Input:", n2)
print("Nth Ugly Number:", nth_ugly_number(n2))
print()
```

```python
import collections

def topKFrequent(words, k):
    word_count = collections.Counter(words)

    def custom_sort(word):
        return (-word_count[word], word)

    sorted_words = sorted(word_count.keys(), key=custom_sort)

    return sorted_words[:k]


words1 = ["i", "love", "leetcode", "i", "love", "coding"]
k1 = 2
print("Input:", words1, ", k =", k1)
print("Top", k1, "Frequent Words:", topKFrequent(words1, k1))
print()

words2 = ["the", "day", "is", "sunny", "the", "the", "the", "sunny", "is", "is"]
k2 = 4
print("Input:", words2, ", k =", k2)
print("Top", k2, "Frequent Words:", topKFrequent(words2, k2))
print()
```

```python
from collections import deque

def maxSlidingWindow(nums, k):
    window = deque()

    result = []

    for i in range(len(nums)):
        if window and window[0] <= i - k:
            window.popleft()

        while window and nums[window[-1]] <= nums[i]:
            window.pop()

        window.append(i)

        if i >= k - 1:
            result.append(nums[window[0]])

    return result


nums1 = [1, 3, -1, -3, 5, 3, 6, 7]
k1 = 3
print("Input:", nums1, ", k =", k1)
print("Max Sliding Window:", maxSlidingWindow(nums1, k1))
print()

nums2 = [1]
k2 = 1
print("Input:", nums2, ", k =", k2)
print("Max Sliding Window:", maxSlidingWindow(nums2, k2))
print()
```

```python
def findClosestElements(arr, k, x):
    left = 0
    right = len(arr) - k

    while left < right:
        mid = (left + right) // 2
        if x - arr[mid] > arr[mid + k] - x:
            left = mid + 1
        else:
            right = mid

    return arr[left:left + k]


arr1 = [1, 2, 3, 4, 5]
k1 = 4
x1 = 3
print("Input:", arr1, ", k =", k1, ", x =", x1)
print("K Closest Elements:", findClosestElements(arr1, k1, x1))
print()

arr2 = [1, 2, 3, 4, 5]
k2 = 4
x2 = -1
print("Input:", arr2, ", k =", k2, ", x =", x2)
print("K Closest Elements:", findClosestElements(arr2, k2, x2))
print()
```
